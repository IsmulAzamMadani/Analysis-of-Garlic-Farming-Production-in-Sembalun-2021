## **Correlation Analysis**

```python
#Select data for correlation analysis
data = df.iloc[:,8:14]
```

```python
data.head()
```

```python
#analize Production data using correlation
correlation_matrix = data.corr()[['Production(Kg)']]
correlation_matrix
```

```python
#Visualize the correlation analysis using heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

from heat map above, we can conclude several things :


*   All of the inputs have a positive correlation to garlic production. It means that if one of the inputs increases, the garlic prosuction tends to increase as well, and vice versa.
*   The highest correlation values is between the usage of garlic Seed and its production.
*  there are two inputs that have a strong correlation to garlic production. Those are: Seed and Phonska
* There are two inputs that have a Moderate correlation to garlic production. Those are: Organic and Urea Fertilizer
* ZA fertilizer is the only input that have a weak correlation to garlic production.



Self Note*:

0-0.3 = very weak correlation

0.3-0.5 = weak correlation

0.5-0.7 = moderate correlation

0.7-0.9 = Strong Correlation
